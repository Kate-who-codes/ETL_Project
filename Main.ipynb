```python
# !pip install psycog2
import pandas as pd
# import psycog2
import getpass
from sqlalchemy import create_engine
```

### Store CSV into Data Frame

```python
PATH_Nys_Math = "./Data/nys_math_test.csv"
nys_math_test_df = pd.read_csv(PATH_Nys_Math)
nys_math_test_df.head()

```

### Convert CSV to Data Frame

```python
PATH_District_Math = "./Data/district_math_test.csv"
district_math_test_df = pd.read_csv(PATH_District_Math)
district_math_test_df.head()
```

### Rename a column in a Data Frame

```python
nys_math_test_df.rename(columns={'Category' : 'Demographic'},
                       inplace=True)
nys_math_test_df.head()
```

### Merge the two data frames


```python
new_df = pd.merge(nys_math_test_df, district_math_test_df, on="Grade")
new_df
```

### Connect to local database

```python
password = ''
while password == '':
    password = getpass.getpass('Password: ')

rds_connection_string = f"postgres:{password}@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
new_df.to_sql(name='merged_table', con=engine, if_exists='append', index=False)
```

### Confirm data has been added

```python
pd.read_sql_query('select * from merged_table', con=engine).head()
```
